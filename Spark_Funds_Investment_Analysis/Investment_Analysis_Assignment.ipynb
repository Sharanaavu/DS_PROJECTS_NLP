```python
import numpy as np
import pandas as pd
```

# Checkpoint 1: Data Cleaning 1

Load companies data into data frame named companies

```python
companies = pd.read_csv("companies.txt", sep="\t")
companies['permalink'] = companies.permalink.str.lower()
companies['homepage_url'] = companies.homepage_url.str.lower()
companies['category_list'] = companies.category_list.str.upper()
companies['status'] = companies.status.str.lower()
companies['country_code'] = companies.country_code.str.upper()
companies['state_code'] = companies.state_code.str.upper()
companies['region'] = companies.region.str.upper()
companies['city'] = companies.city.str.upper()
companies.head()
```

Load rounds data into data frame named rounds2

```python
rounds2 = pd.read_csv("rounds2.csv", encoding="ISO-8859-1")
rounds2['company_permalink'] = rounds2.company_permalink.str.lower()
rounds2['funding_round_permalink'] = rounds2.funding_round_permalink.str.lower()
rounds2['funding_round_type'] = rounds2.funding_round_type.str.lower()
rounds2['funding_round_code'] = rounds2.funding_round_code.str.upper()
rounds2.head()
```

```python
# unique companies present in rounds2
print(len(rounds2.company_permalink.unique()))
```

```python
# unique companies present in companies
print(len(companies.permalink.unique()))
```

```python
# permalink column can be used as the unique key for each company
companies.describe()
```

```python
# companies in the rounds2 file which are not present in companies
rounds2_permalink = set(rounds2.company_permalink)
companies_permalink = set(companies.permalink)
permalink_in_rounds2_but_not_in_companies = rounds2_permalink - companies_permalink
print(len(permalink_in_rounds2_but_not_in_companies))
```

Merge comapnies and rounds2 data frames so that all variables (columns) in the companies frame are added to the rounds2 data frame. Name the merged frame master_frame.

```python
master_frame = pd.merge(companies, \
                        rounds2.rename(columns={'company_permalink':'permalink'}), \
                        how='outer', \
                        on='permalink')
master_frame.shape
```

# Checkpoint 2: Funding Type Analysis

This is the first of the three goals of data analysis – investment type analysis.

The funding types such as seed, venture, angel, etc. depend on the type of the company (startup, corporate, etc.), its stage (early stage startup, funded startup, etc.), the amount of funding (a few million USD to a billion USD), and so on. For example, seed, angel and venture are three common stages of startup funding.

- Seed/angel funding refer to early stage startups whereas venture funding occurs after seed or angel stage/s and involves a relatively higher amount of investment.
- Private equity type investments are associated with much larger companies and involve much higher investments than venture type. Startups which have grown in scale may also receive private equity funding. This means that if a company has reached the venture stage, it would have already passed through the angel or seed stage/s.

Spark Funds wants to choose one of these four investment types for each potential investment they will make.

Considering the constraints of Spark Funds, decide one funding type which is most suitable for them.

1. Calculate the most representative value of the investment amount for each of the four funding types (venture, angel, seed, and private equity)
2. Based on the most representative investment amount calculated above, find the most suitable investment type for Spark Funds?

Considering that Spark Funds wants to invest between **5 to 15 million USD** per investment round, which investment type is the most suitable for it? Identify the investment type and, for further analysis, filter the data so it only contains the chosen investment type.

```python
# calculate avergage funding amount for investment types'venture','angel','seed','private_equity'
mean_fund_raised_per_investment_type = master_frame[master_frame['funding_round_type'].isin(['venture','angel','seed','private_equity'])] \
        .groupby('funding_round_type') \
        .agg({'raised_amount_usd':'mean'}) \
        .rename(columns={'raised_amount_usd':'mean_raised_amount_usd'}) \
        .sort_values(by='mean_raised_amount_usd',ascending=False)
mean_fund_raised_per_investment_type['mean_raised_amount_usd'].map('{:,.2f}'.format)
```

Since investment type __venture__ is 11 million, it lies in the investment range of interest

### Selected investment type: venture

# Checkpoint 3: Country Analysis

```python
# For investment type venture, make a data frame named top9 with the top nine countries (based on the total investment amount each country has received)
top9 = master_frame[master_frame['funding_round_type'] == 'venture'] \
                                .groupby('country_code')['raised_amount_usd'] \
                                .sum() \
                                .reset_index(name='sum') \
                                .rename(columns={'sum':'total_investment_usd'}) \
                                .sort_values(by='total_investment_usd',ascending=False) \
                                .head(9)
top9
```

# Checkpoint 4: Sector Analysis 1

```python
# Extract the primary sector of each category list from the category_list column
master_frame['primary_sector'] = master_frame.category_list.str.split('|').str[0]
```

```python
# Load mapping file mapping.csv into a dataframe
mapping = pd.read_csv("mapping.csv")
mapping['category_list'] = mapping.category_list.str.upper()
mapping.head()
```

```python
# map each primary sector to one of the eight main sectors
# the sector names in category_list had an error wherein NA was replaced by 0. e.g. ANALYTICS was written as A0LYTICS
mapping["category_list"] = mapping["category_list"].str.replace(r"(?<=\w)0(?=\w)", "NA")
mapping['main_sector'] = mapping[['Automotive & Sports', \
                             'Blanks', \
                             'Cleantech / Semiconductors', \
                             'Entertainment', \
                             'Health', \
                             'Manufacturing', \
                             'News, Search and Messaging', \
                             'Others', \
                             'Social, Finance, Analytics, Advertising' \
                            ]].idxmax(axis=1)
mapping[['category_list','main_sector']]
```

```python
# merged data frame with each primary sector mapped to its main sector 
master_frame_with_sector = pd.merge(master_frame, mapping[['category_list','main_sector']].rename(columns={'category_list':'primary_sector'}), how='left', on='primary_sector')
master_frame_with_sector.head()
```

# Checkpoint 5: Sector Analysis 2

Find out the most heavily invested main sectors in each of the three countries (for funding type **venture** and investments range of **5-15 M USD**)

## United States

```python
# D1 data frame for United States 
D1 = master_frame_with_sector[(master_frame_with_sector['country_code']=='USA') & \
                              (master_frame_with_sector['funding_round_type']=='venture') & \
                              (master_frame_with_sector['raised_amount_usd'] >= 5000000) & \
                              (master_frame_with_sector['raised_amount_usd'] <= 15000000)]

# total number (or count) of investments for each main sector
investment_count_by_sector_D1 = D1.groupby('main_sector') \
                                .size() \
                                .reset_index(name='size') \
                                .rename(columns={'size':'count_of_investments'})

# total amount invested in each main sector
investment_sum_by_sector_D1 = D1.groupby('main_sector')['raised_amount_usd'] \
                                .sum() \
                                .reset_index(name='sum') \
                                .rename(columns={'sum':'sum_of_investments'})

# merge total number (or count) of investments and total amount invested for each main sector into D1
investment_stats_by_sector_D1 = pd.merge(investment_count_by_sector_D1, investment_sum_by_sector_D1, how='inner', on='main_sector')
D1 = pd.merge(D1, investment_stats_by_sector_D1, how='left', on='main_sector')
D1['sum_of_investments'] = D1['sum_of_investments'].map('{:,.2f}'.format)
D1.head()
```

```python
# Total number and amount of investments in United States
total_number_of_investment_D1 = investment_stats_by_sector_D1[['count_of_investments']].sum()
total_amount_of_investment_D1 = investment_stats_by_sector_D1[['sum_of_investments']].sum().map('{:,.2f}'.format)
print(total_number_of_investment_D1, total_amount_of_investment_D1)
```

```python
# Top 3 sector (based on count of investments) in United States
top_3_sectors_D1 = investment_stats_by_sector_D1.sort_values(['count_of_investments', 'sum_of_investments'], ascending=[False, False]).head(3)
top_3_sectors_D1
```

```python
# Companies which received the highest investment in each of top 3 sector in United States
D1_without_fund_range = master_frame_with_sector[(master_frame_with_sector['country_code']=='USA') & \
                                                 (master_frame_with_sector['funding_round_type']=='venture') & \
                                                 (master_frame_with_sector['main_sector'].isin(top_3_sectors_D1['main_sector']))]
idx_D1 = D1_without_fund_range.groupby(['main_sector'])['raised_amount_usd'].transform(max) == D1_without_fund_range['raised_amount_usd']
D1_without_fund_range[idx_D1][['main_sector','name']]
```

# Checkpoint 6: Plots

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,10))

data = master_frame[master_frame['funding_round_type'].isin(['venture','seed','private_equity'])]

# subplot 1
plt.subplot(1, 2, 1)
plt.title('Investment count on linear scale')
ax1 = sns.barplot(x='funding_round_type', y='raised_amount_usd', data=data, estimator=np.size)
ax1.set(xlabel='Investment type', ylabel='Number of investments')

# subplot 2
plt.subplot(1, 2, 2)
plt.title('Investment value on logarithmic scale')
ax2 = sns.barplot(x='funding_round_type', y='raised_amount_usd', data=data)
ax2.figure.get_axes()[1].set_yscale('log')
ax2.set(xlabel='Investment type', ylabel='Average investment in USD')

plt.show()
```

```python
plt.figure(figsize=(10,10))

plt.title('Top countries by investment in venture funds on logarithmic scale')
ax = sns.barplot(x='country_code', \
            y='total_investment_usd', \
            data=top9)
ax.figure.get_axes()[0].set_yscale('log')
ax.set(xlabel='Country', ylabel='Total investment in USD')

plt.show()
```
